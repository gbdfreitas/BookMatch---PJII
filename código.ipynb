```python
# Importando as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuração para exibir gráficos no Jupyter
%matplotlib inline

# Carregar os dados (Substitua o caminho pelo correto)
df = pd.read_csv('kindle_data-v2.csv')

# Visualizar as primeiras linhas do dataset
print("Primeiras linhas do dataset:")
print(df.head())

# Verificar o formato dos dados
print("\nInformações sobre o dataset:")
print(df.info())

# Resumo estatístico das variáveis numéricas
print("\nResumo estatístico das variáveis numéricas:")
print(df.describe())

# Verificar o número de valores ausentes por coluna
print("\nValores ausentes por coluna:")
print(df.isnull().sum())

# Análise da variável 'price' (Preço dos livros)
plt.figure(figsize=(10,6))
sns.histplot(df['price'], kde=True, color='blue')
plt.title('Distribuição dos Preços dos Livros')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.show()

# Análise da variável 'stars' (Avaliações dos livros)
plt.figure(figsize=(10,6))
sns.histplot(df['stars'], bins=10, kde=True, color='green')
plt.title('Distribuição das Avaliações dos Livros')
plt.xlabel('Classificação (Estrelas)')
plt.ylabel('Frequência')
plt.show()

# Verificar a distribuição do número de reviews
plt.figure(figsize=(10,6))
sns.histplot(df['reviews'], kde=True, color='purple')
plt.title('Distribuição do Número de Avaliações (Reviews)')
plt.xlabel('Número de Avaliações')
plt.ylabel('Frequência')
plt.show()

# Boxplot do preço dos livros vendidos por diferentes vendedores
plt.figure(figsize=(10,6))
sns.boxplot(x='soldBy', y='price', data=df)
plt.title('Distribuição dos Preços por Vendedor')
plt.xlabel('Vendedor')
plt.ylabel('Preço')
plt.xticks(rotation=90)
plt.show()

# Verificar correlações entre variáveis numéricas
correlacao = df[['price', 'stars', 'reviews']].corr()
print("\nMatriz de correlação entre variáveis numéricas:")
print(correlacao)

# Visualizar a matriz de correlação
plt.figure(figsize=(8,6))
sns.heatmap(correlacao, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlação entre Preço, Avaliação e Reviews')
plt.show()

# Scatterplot: Avaliações (stars) vs Preço
plt.figure(figsize=(10,6))
sns.scatterplot(x='price', y='stars', data=df)
plt.title('Correlação entre Avaliação (Estrelas) e Preço')
plt.xlabel('Preço')
plt.ylabel('Classificação (Estrelas)')
plt.show()

# Análise da variável 'isKindleUnlimited'
plt.figure(figsize=(10,6))
sns.countplot(x='isKindleUnlimited', data=df)
plt.title('Distribuição dos Livros no Kindle Unlimited')
plt.xlabel('Disponível no Kindle Unlimited')
plt.ylabel('Frequência')
plt.show()
```

```python
# Remover outliers de preço (exemplo removendo valores acima de 500)
df = df[df['price'] < 500]
```

```python
# Criar feature para popularidade do autor
df['author_popularity'] = df.groupby('author')['stars'].transform('mean')
```
